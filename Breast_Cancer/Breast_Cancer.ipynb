```python
df = pd.read_csv('gdrive/MyDrive/Datasets/Breast_Cancer/data.csv')
```

```python
#preprocessing
df.dropna(inplace=True,axis=1) #drop all NaNs
df.drop('id',axis=1,inplace=True) #drop id column as it is of no use
labels = df['diagnosis'] #store labels in a different dataframe
df.drop('diagnosis',axis=1,inplace=True) #drop labels columns from the original dataframe
df.head()
```

```python
#convert the dataframes into numpy arrays
data = df.to_numpy()
labels = labels.to_numpy()
for i in range(len(labels)):
  if(labels[i]=='M'):
    labels[i] = 1
  else:
    labels[i] = 0
```

```python
#split the data into train and test sets
X_train,X_test,Y_train,Y_test = train_test_split(data,labels,test_size=0.2)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train = X_train.T
X_test = X_test.T
Y_train = Y_train.reshape(1,455)
Y_test = Y_test.reshape(1,114)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
W = np.random.randn(30,1)*0.01
b = 0
```

```python
#Hyperparameters
num_epochs = 5000
lr = 0.001
```

```python
def Cost(y,y_hat):
  return -(np.multiply(y,np.log(y_hat))+np.multiply(1-y,np.log(1-y_hat)))
```

```python
def sigmoid(z):
  z = np.array(z,dtype=np.float32)
  return 1/(1+np.exp(-z))
```

```python
cost = []
iter = []
#train loop
for i in range(num_epochs):
  z = np.dot(W.T,X_train)+b
  # print(z)
  a = sigmoid(z)
  c = np.sum(Cost(Y_train,a))
  cost.append(c)
  iter.append(i)
  dz = a-Y_train
  dw = (1/455)*(np.dot(X_train,dz.T))
  db = (1/455)*(np.sum(dz))

  W = W-lr*dw
  b = b-lr*db


```

```python
def check_accuracy(x,y):
  count = 0
  total = 0
  y_hat = sigmoid(np.dot(W.T,x))
  predictions = np.array([1 if(y_hat[0][i]>=0.5) else 0 for i in range(y_hat.shape[1])])
  for i in range(y_hat.shape[1]):
    if(y[0][i]==predictions[i]):
      count+=1
    total+=1

  return count/total
```

```python
print(check_accuracy(X_train,Y_train))
print(check_accuracy(X_test,Y_test))
```
